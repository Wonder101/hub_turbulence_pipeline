## Step 2: Haversine Distance (function) calculation
Contextually: The Earth is a sphere, so we can't measure the distance between two lat/lon points with flat geometry. Haversine computes the correct curved-surface distance in miles. We wrap it in a function so we can reuse it for every hub without copy-pasting the math five times. 

My data has raw lat/long. I want to know how far each report is from ATL, ORD, FJK, etc. 

- The key idea to look for: instead of looping row by row, you pass entire NumPy arrays (your whole latitude column, your whole longitude column) into np.sin, np.cos, np.arctan2 all at once. This is called vectorization: it's the core reason NumPy exists.
Haversine distance between two points on a sphere needs four ingredients:
1. Both latitudes and longitudes converted from degrees to radians (trig functions in NumPy expect radians, not degrees)
2. The difference in latitude and the difference in longitude between your two points
3. A formula combining sines and cosines of those differences
4. Multiplying the result by Earth's radius to get an actual distance (miles or km)

"a" term: 
a = sin²(Δlat/2) + cos(lat1) · cos(lat2) · sin²(Δlon/2)
This term has two pieces added together: one for north south separation and one for east west separation. 
- sin²(Δlat/2): measures how far apart the two points are in latitude.

- sin²(Δlon/2): measures how far apart they are in longitude, but scaled down by the cosines *(cos(lat1) · cos(lat2))*.
The scaling matters because *lines of longitude squeeze together* as you move toward the poles (a degree of longitude near the equator covers more ground than a degree of longitude up near canada) The cosine terms correct for that squeeze .
- a represents: the haversine of the *central angle*

"C" term: c = 2 · arctan2(√a, √(1-a)) Once we have a we need to convert it back into a an actual angle in radians (Recall that the central angle is the literal angle between points A and B measured from Earth's center.) 
- *arctan2(x1,x2)* is used instead of the simpler arcsin beause it stays numerically stable even for points that are very close together or nearly on opposite sides of the planet
- arctan = given opposite and adjacent, you can find the angle that lies on the adjcaent across the opposite
- Why is arctan2 more stable than arcsin?: the derivative (slope) of arcsin(x) is d/dx[arcsin(x)] = 1 / √(1 - x²), as x approaches 1 the denominator approaches 0, which means the slope shoots toward infinity. Near x =1, the curve goes almost vertical.
- Computers don't store numbers with inifinite precision, there's always a tiny rounding error somewhere in the last decimal place. When we feed a number into a function with a near vertical slope, that rounding error gets amplified into a much bigger error in the output. A 0.0000001 error in our input might turn into a 0.01 error in your angle. 
distance = R * c is basic circle geometry: arc length = radius * angle (in radians) 

## Step 3: Filter down to reports near a hub
Contextually: We define each hub as a lat/lon coordinate pair. The loop runs haversine once per hub, each run is vectorized: it computes the distance from that one hub to all ~5000 report locations simultaneously, adding one new column per hub to the dataframe.

Right now we have a distance column for ATL, let's extend that to multple hubs and combine them into one filter. We can wrap our haversine math into a function.

## Step 4: Filter to reports near any hub 
Contextually: We build a boolean mask (a column of T/F values, one per row) which marks any report within 100 miles of at least one hub as True. Then we use that mask to slice the dataframe down to only those relevant rows. **.copy()** is important here, it allows us to make this sliced dataframe a whole independent table (not a window to the original) so future writes don't trigger warnings.

```python
# Step A: Turn haversine code into a function 

def haversine(lat1, lon1, lat2, lon2): 
    R = 3956 
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2]) 
    dlat = lat2 - lat1 
    dlon = lon2 - lon1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2 
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a)) 
    return R * c 

# Step B: Define hub coordinates 
hubs = {
    'ATL': (33.6407, -84.4277), 
    'ORD': (41.9786, -87.9048), 
    'JFK': (40.6399, -73.7787), 
    'LAX': (33.9425, -118.4072), 
    'DFW': (32.897, -97.038), 
}

# Step C: Loop through hubs, create one distance column per hub 
for name, (hub_lat, hub_lon) in hubs.items(): 
    ds[f'dist_{name}'] = haversine(hub_lat, hub_lon, ds['latitude'], ds['longitude']) 

# Step D: Build mask; is this report near any hib? 
threshold = 100 # 100 mi radius threshold 
near_hub_mask = ((ds['dist_ATL'] < threshold) | (ds['dist_ORD'] < threshold) | (ds['dist_JFK'] < threshold) | (ds['dist_LAX'] < threshold) | (ds['dist_DFW'] < threshold))
near_ds = ds[near_hub_mask].copy() # THIS .copy() IS IMPORTANT DOWN THE LINE 

near_ds.shape #filtered where near_hub_mask is True
```

## Step 5: Turn raw columns into one single number per row (risk score)
Contextually: **np.select** is a vectorized if/elif/else that checks every row against your conditions list in order (position 0 then position 1...), assigns the matching choice and falls back to the default if nothing matches. You put the most severe condition first because the first match wins. NaN rows are explicitly caught with **pd.isna()** and scored as unknown rather than assuming they're safe.

```python
# What are we working with? Looking for Turbulence.
print(near_ds.columns)
```

```python
# Step A: Find columns related to turbulence and icing
print(near_ds['turbulence_intensity'].unique())

# This matters because conditions we write depend entirely on whether that column holds numbers (0,1..) or text codes ('LGT', 'MOD'..) You can't write the condition correctly until you know which one it is.
```

```python
# Step B: Define conditions, ranked most severe first; 
# NaN: Pilot filed report but didn't include turbulence data at all 
conditions = [
    near_ds['turbulence_intensity'] == 'MOD', # Moderate, position 0  
    near_ds['turbulence_intensity'] == 'LGT', # Light, position 1
    near_ds['turbulence_intensity'] == 'NEG', # Negative or None, position 2 
    pd.isna(near_ds['turbulence_intensity']) == 'NAN' # Not a number, position 3 
] 

choices = ['Moderate', 'Light', 'Negative or None', 'Not a Number'] # Pos 0, pos 1, pos 2, pos 3

# Step C: Apply 
near_ds['risk_score'] = np.select(conditions, choices, default = 'EDGE-CASE')
near_ds['risk_score'].sample(10)

# Go through all of conditions, enforce the choice list if it fufills the condition (in the event of ties take the first condition), default if the value doesn't fufill any of the conditions. 
#Issue: Previously I didn't have .copy() for near_ds and it was telling me that a value is trying to be set on a copy of a slice from a dataframe. This is because near_ds was created by filtering ds with a boolean mask. 
#Pandas isn't sure whether near_ds is a brand new independent table or just a window pointing at the original ds in memory (like views in sql.)
```

## Step 6: Import into PostgreSQL 
Contextually: Right now, every time we restart the notebook, all the data vanishes. It only exists in memory while Python is running. PostgreSQL gives it a permanent home. Later, oir Streamlit dashboard will read from that database instead of re-downloading and re-processing the whole FAA feed every time someone looks at it.

We'll need two libraries: 
1. sqlalchemy: handles the translation between python and (any) SQL. Also handles data sockets.
2. psycopg2 (in our situation psycopg2-binary): while sqlachemy does the translation psycopg2 does the actual transmission between the two. For our situation, we download psycopg2-binary since everything is already precompiled and easy to use. Our OS kernel will send bytes into its outbound buffer, it'll be packaged and sent to the inbound buffer managed by the same kernel (since we're running this locally.) The package will then be reconstructed and anything need will be sent back.

```python
# Step A: Create our table in dbeaver first (this means create table statement, we also can use near_ds.info() to see data-types.
# Note: The difference between a text datatype and a varchar. 

# Step B: Build the connection in Python 
# engine = object SQLAlchemy created from create_engine(..), we can think of it as holding information telling where Postgres is and info to log in, a connection object. Nothing moving yet though.
from sqlalchemy import create_engine 
engine = create_engine('postgresql+psycopg2://postgres:8319@localhost:5432/weather_proj')

# We're renaming columns in df before saving because PostgreSQL lowercases everything. 
# When pandas sends the data over it posses the column name exactly as it exists in the df. 
# However dist_ATL has capital letters, then postgreSQL looks for a column literally named dist_ATL 
# and can't find it because it stored as dist_atl 
near_ds = near_ds.rename(columns={
    'dist_ATL': 'dist_atl', 
    'dist_ORD': 'dist_ord', 
    'dist_JFK': 'dist_jfk', 
    'dist_LAX': 'dist_lax', 
    'dist_DFW': 'dist_dfw'
}) 

# Step C: Push our dataframe 
cols_to_save = [
    'receipt_time', 'observation_time', 'aircraft_ref', 'latitude', 
    'longitude', 'turbulence_intensity', 'risk_score', 'dist_atl', 
    'dist_ord', 'dist_jfk', 'dist_lax', 'dist_dfw'
]

#Pandas took the filtered, scored dataframe, handed it to SQLAlchemy via con=engine, which translated it into SQL INSERT statements and sent them to PostgreSQL. 
#The 19 it returned confirmed 19 rows landed successfully.
#if_exists='append' means future runs add new rows rather than wiping history. 
#index=False stopped pandas from writing its internal row numbers as a meaningless extra column.
near_ds[cols_to_save].to_sql(
    name='hub_turbulence_reports', 
    con=engine,
    if_exists='append',
    index=False 
) 
```
